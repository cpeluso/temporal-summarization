### Producing results

```python
# Questo blocco di codice permette di creare un file di run per una singola query.
# La query è definita dalla variabile query_id.
#
# Nella funzione make_run(query_id), 
# il modello finetunato viene scaricato e viene fatta inference sui testi associati alla query.
# Dunque, i risultati dell'inference vengono formattati in un file run-<query_id>.txt
# che verrà usato successivamente dallo script di valutazione (non in questo blocco.)

# Nella funzione make_runs(), invece,
# il modello finetunato viene scaricato e viene fatta inference sui testi associati a ciascuna query.
# Dunque, i risultati dell'inference vengono formattati in N file run-<query_id>.txt
# che verrà usato successivamente dallo script di valutazione (non in questo blocco.)
# ATTENZIONE: la funzione make_runs() impiega molto tempo a terminare
# perchè viene fatta inference su tutti i dati.
# Nota che non è necessario ri-fare inference, in quanto i file run-<query_id>.txt
# sono già salvati nella cartella results della repo GitHub.

from trec.trec_evaluator import TrecEvaluator

query_id = 1

trec_evaluator = TrecEvaluator()

# If needed to make a run for a single query
trec_evaluator.make_run(query_id)

# If needed to make a run for every query (could take up to 2 hours):
#trec_evaluator.make_runs()
```

### Evaluating single result (at least 3 runs are needed)

```python
# In questo blocco, i file run-<query_id>.txt vengono usati dallo script di valutazione
# proposto dal TREC (funzione evaluation_script(*args)).
# In output vengono visualizzati i risultati.
#
# Questo blocco è a scopo di esempio e valuta solamente le prime 3 run.

from src.data.s3_connector import download_file
from trec.eval import evaluation_script
import os

if not os.path.exists("data"):
  os.makedirs("data")

nuggets_filename = "data/nuggets.tsv"
updates_filename = "data/updates.tsv"
matches_filename = "data/matches.tsv"

if not os.path.exists(nuggets_filename):
  download_file(nuggets_filename)

if not os.path.exists(updates_filename):
  download_file(updates_filename)

if not os.path.exists(matches_filename):
  download_file(matches_filename)

runs_filename = ["./results/run-1.txt", "./results/run-2.txt", "./results/run-3.txt"] 

evaluation_script(
    nuggets_filename,
    updates_filename,
    runs_filename,
    matches_filename,
    None,
    args_debug     = False,
    args_binaryrel = False
)
```
